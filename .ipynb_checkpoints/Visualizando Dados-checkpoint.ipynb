# Gráfico de Barra

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
```

```python
# barras possuem o tamanho padrão de 0.8, então adicionaremos 0.1 às 
# coordenadas à esquerda para que cada barra seja centralizada
xs = [i + 0.5 for i, _ in enumerate(movies)]
```

```python
# as barras do gráfico com as coordenadas x à esquerda [xs], alturas [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# de Premiações")
plt.xlabel("Meus Filmes Favoritos")

# nomeia o eixo x com nomes de filmes na barra lateral
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
# Arredonda a casa decimal - ex: 12 = 10 / 18 = 10 ...
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)
```

```python
plt.bar([x for x in histogram.keys()], histogram.values(), 8)
```

```python
plt.axis([-5, 105, 0, 5])
```

```python
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decil")
plt.ylabel("# de Alunos")
plt.title("Distribuição das Notas do Teste 1")
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]
```

```python
plt.bar([2012.6, 2013.6], mentions, 0.8)
plt.xticks(years)
plt.ylabel("# de vezes que ouvimos alguém dizer 'data science'")
```

```python
# e então adiciona a +2.013e3 para fora do canto(matplotlib feio!)
plt.ticklabel_format(useOffset=False)
plt.bar([2012.6, 2013.6], mentions)
plt.xticks(years)
# enganar o eixo y mostra apenas a parte acima de 500
plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Olhe o 'Grande' Aumento!")
plt.show()
```

# Gráfico de Linhas  


Como já haviamos dito, podemos construir gráficos de linha usando plt.plot(). Eles são uma boa escolha para mostrar tendências.

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
# podemos fazer multiplas chamadas para plt.plot
# para mostrar multiplas séries no mesmo gráfico
plt.plot(xs, variance, 'g-', label='variance') # linha verde sólido
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # linha com linha de ponto tracejado vermelho
plt.plot(xs, total_error, 'b:', label='total error') # linha com pontilhado azul

# porque atribuímos rótulos para cada série
# podemos obter uma legenda gratuita
# loc=9 significa "top center"
plt.legend(loc=9)
plt.xlabel("complexidade do modelo")
plt.title('Compromisso entre Polarização e Variância')
plt.show()
```

# Gráfico de Dispersão

Um gráfico de dispersão é a escolha certa para visualizar o relacionamento entre dois pares de conjuntos de dados. 

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)
```

```python
# nomeia cada posição
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy=(friend_count, minute_count), # coloca o rótulo com sua posição
                xytext=(5, -5),                  # mas compensa um pouco
                textcoords='offset points')

plt.title("Minutos Diários vs. Número de Amigos")
plt.xlabel("# de Amigos")
plt.ylabel("minutos diários passados no site")
plt.scatter(friends, minutes)
```

> Nota: se você está espalhando variáveis comparáveis, talvez você obtenha uma imagem enganosa se deixar matplotlib escolher a escala, como podemos visualizar no exemplo abaixo:

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
```

```python
plt.scatter(test_1_grades, test_2_grades)
plt.title('Os eixos não são compatíveis')
plt.xlabel('nota de teste 2')
plt.ylabel('nota de teste 1')
plt.show()
```

Se incluirmos uma chamada para **plt.axis("equals")**, o gráfico será mais preciso conforme exemplo abaixo:

```python
plt.scatter(test_1_grades, test_2_grades)
plt.title('Os eixos não são compatíveis')
plt.xlabel('nota de teste 2')
plt.ylabel('nota de teste 1')
plt.axis("on")
plt.show()
```
